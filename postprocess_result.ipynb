```python
import pandas as pd
import csv
```

```python
result_fname = 'result_gpt.csv'
```

```python
df = pd.read_csv(result_fname)
df
```

```python
# Group by the "style" column and filter for those with more than 10 correct & incorrect samples
grouped_df = df.groupby(['style']).filter(lambda x: True)
```

```python
grouped_summary = grouped_df.groupby(['style']).size().reset_index(name='counts')
grouped_summary
```

```python
sorted_summary = grouped_summary.sort_values(by='counts', ascending=False)
sorted_summary
```
